```python
import matplotlib.pyplot as plt

# Array lengths (same for all algorithms)
array_lengths = [10, 100, 1000, 10000]

# Time data for each algorithm (in seconds)
insertion_sort_times = [1.06e-06, 2.689e-05, 0.0016113, 0.150568]
bubble_sort_times = [9.29e-07, 6.43e-05, 0.00457804, 0.45905]
optimized_bubble_sort_times = [8.9e-07, 6.54e-05, 0.00447369, 0.454065]
selection_sort_times = [9.2e-07, 4.1931e-05, 0.00301411, 0.243591]

# Create a plot
plt.figure(figsize=(10, 6))

# Plot each sorting algorithm's time vs array length
plt.plot(array_lengths, insertion_sort_times, label="Insertion Sort", marker='o', linestyle='-', color='blue')
plt.plot(array_lengths, bubble_sort_times, label="Bubble Sort", marker='o', linestyle='-', color='green')
plt.plot(array_lengths, optimized_bubble_sort_times, label="Optimized Bubble Sort", marker='o', linestyle='-', color='orange')
plt.plot(array_lengths, selection_sort_times, label="Selection Sort", marker='o', linestyle='-', color='red')

# Label the axes and the title
plt.xlabel('Array Length')
plt.ylabel('Time Taken (seconds)')
plt.title('Time Comparison of Sorting Algorithms')

# Set logarithmic scale for better visualization of time differences
plt.yscale('log')

# Add a legend
plt.legend()

# Display the plot
plt.grid(True, which="both", ls="--")
plt.show()
```
